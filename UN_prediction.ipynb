```python
df = pd.read_csv('data/TrainingSet.csv',index_col=0)
sub_lbl = pd.read_csv('SubmissionRows.csv',index_col=0)
df.head()
```

```python
def generate_year_list(start, stop=None):
    """ 
    make a list of column names for specific years
    in the format they appear in the data frame start/stop inclusive
    """  
    if isinstance(start, list):
        data_range = start
    elif stop:
        data_range = range(start, stop+1)
    else:
        data_range = [start]
    
    yrs = []
    
    for yr in data_range:
        yrs.append("{0} [YR{0}]".format(yr))
        
    return yrs

# ========== TEST CASES =======
# one year
print generate_year_list(2008)

# start and stop (inclusive)
print generate_year_list(1985, 1990)

# custom year list
print generate_year_list([1985, 1990])
```

```python
prediction_rows = df.loc[sub_lbl.index]
prediction_rows.head()
```

```python
import statsmodels.graphics.tsaplots as tsaplots
prediction_data = prediction_rows[generate_year_list(1972,2007)]
prediction_data.head()
```

```python
# grab a random sample of 10 of the timeseries
np.random.seed(896)
rand_rows = np.random.choice(prediction_rows.index.values, size=10)

def plot_rows(data, ids=None, linestyle="-", legend=True):
    # get some colors for the lines
    bmap = pplt.brewer2mpl.get_map('Set3','Qualitative', 10)
    colors = bmap.mpl_colors
    
    if not None == ids:
        get_rows = lambda: enumerate(ids)
    else:
        get_rows = lambda: enumerate(data.index.values)
    
    for i, r in get_rows():
        # get the time series values
        time_data = data.loc[r]

        # create an x axis to plot along
        just_years = [y[:4] for y in data.columns]
        X = pd.DatetimeIndex(just_years)

        # get time series info for labeling
        country, descrip = df[["Country Name", "Series Name"]].loc[r]

        # plot the series
        plt.plot(X, time_data, c=colors[i],
                 label="{} - {}".format(country, descrip), ls=linestyle)
        plt.scatter(X, time_data, alpha=0.8, c=colors[i])

    if legend:
        plt.legend(loc=0)
    plt.title("Progress Towards Subset of MDGs")

plot_rows(prediction_data, ids=rand_rows)
plt.show()
```

```python
# find first yr, in aa given row, that is not NaN
def find_first_yr(sr):
    return int(sr.first_valid_index().split(' ')[0])

# time in years since 1972
def generate_year_dist(yr):
    return yr - 1972

# test cases
X=[]
for i in range(1972,2008):
    X.append(generate_year_dist(i))
print X

# test case for find_first_yr
print find_first_yr(prediction_data.loc[559])
```

```python
rows = prediction_data.iterrows()
with open('SubmissionRows1.csv','wb') as out:
    write = csv.writer(out,delimiter=',')
    write.writerow(['', '2008 [YR2008]', '2012 [YR2012]'])
    # create a numpy array of years(in dist from 1972), and numpy array of recorded progress towards UN MGDs'
    for i,row in rows:
        first_yr = find_first_yr(row)
        x = []
        for j in range(first_yr,2008):
            x.append(generate_year_dist(j))
        x = x[-3:]
        y = row.loc[row.first_valid_index():'2007 [YR2007]'].values
        x = np.array(x)
        X = np.vstack(x)
        mask = np.isnan(y)
        y[mask] = np.interp(np.flatnonzero(mask),np.flatnonzero(~mask),y[~mask])
        y = y[-3:]
        # lr = LinearRegression() add ??
        # lr.fit(X,y)
        # fit two ridge regression models, second is experimental
        r = Ridge(alpha=.0001)
        r2 = Ridge(alpha=.003,normalize=True,solver='svd')
        r.fit(X,y)
        r2.fit(X,y)
        write.writerow([i,(r.predict([36])+r2.predict([36]))/2,(r.predict([40])+r2.predict([40]))/2])
```
